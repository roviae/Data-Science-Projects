```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
y = raw_data.Profit
x1 = raw_data['Marketing Expenditure']
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```
